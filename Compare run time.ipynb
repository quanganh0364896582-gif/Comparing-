```python
import matplotlib.pyplot as plt
import numpy as np

# 1. Chuẩn bị dữ liệu
labels = ['1', '2', '4', '6', '8', '10', '12', '14', '16', '18']

# Dữ liệu cho DAI
dai_dnn = [219079.9, 223062.1, 224755.9, 222866.8, 228114.1, 223305.3, 229586.2, 227814.6, 224485.1, 227404.2]
dai_si = [13794703.4, 13791338.6, 13789993.9, 13801657.6, 13773531.0, 13889284.4, 13806948.8, 13807672.4, 13829113.5, 13804147.6]

# Dữ liệu cho Local
local_dnn = [527520.0, 497370.0, 367810.0, 427960.0, 501620.0, 330730.0, 329620.0, 454000.0, 500950.0, 394030.0]
local_si = [24370820.0, 25699640.0, 23537930.0, 24760380.0, 23874730.0, 23181130.0, 32885630.0, 26683290.0, 26755150.0, 24643820.0]

dai_dnn_log = np.log10(dai_dnn)
dai_si_log = np.log10(dai_si)
local_dnn_log = np.log10(local_dnn)
local_si_log = np.log10(local_si)

# 2. Thiết lập thông số vẽ
x = np.arange(len(labels))
width = 0.35

# Tạo khung hình 1 hàng 2 cột (do bạn mới gửi dữ liệu 2 máy)
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

# --- Vẽ biểu đồ DAI ---
ax1.bar(x - width/2, dai_dnn_log, width, label='DNN', color='#1f77b4')
ax1.bar(x + width/2, dai_si_log, width, label='SI', color='#ff7f0e')
ax1.set_title('DAI')
ax1.set_ylabel('log(Time ns)')
ax1.set_xlabel('Number of edges')
ax1.set_xticks(x)
ax1.set_xticklabels(labels)
ax1.set_ylim(0, 10) # Giới hạn trục y để các cột rõ ràng
ax1.grid(axis='y', linestyle='--', alpha=0.6)
ax1.legend()

# --- Vẽ biểu đồ Local ---
ax2.bar(x - width/2, local_dnn_log, width, label='DNN', color='#1f77b4')
ax2.bar(x + width/2, local_si_log, width, label='SI', color='#ff7f0e')
ax2.set_title('Local')
ax2.set_ylabel('log(Time ns)')
ax2.set_xlabel('Number of edges')
ax2.set_xticks(x)
ax2.set_xticklabels(labels)
ax2.set_ylim(0, 10)
ax2.grid(axis='y', linestyle='--', alpha=0.6)
ax2.legend()

plt.tight_layout()
plt.show()
```
